```python
## outlier_capping
def outlier_capping(df1,column):
    Q1=df1[column].quantile(0.25)
    Q3=df1[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df1[column]=df1[column].apply(lambda x:lower_bound if x<lower_bound else upper_bound if x>upper_bound else x)
for col in df1.select_dtypes(['int','float']).columns:
    outlier_capping(df1,col)
```

## Model Building:

```python
## Spliting the dataset into feature and target
## then into training and testing data of titanic_test and titanic_train to feed the model
```

```python
## target:
y_train = df1['Survived']
y_train.head()
```

```python
## features:
x_train = df1.drop(columns='Survived')
x_train.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_train,y_train,train_size=0.75,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## Building a model and feeding with training data
```

## Logistic_Regression:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
model = LogisticRegression(multi_class='ovr')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy_score(y_test,y_pred)
```

## Model Evaluation:

#### Performance_matrix

```python
## confusion_matrix:
from sklearn.metrics import confusion_matrix,classification_report
```

```python
conf=confusion_matrix(y_test,y_pred)
conf
```

```python
sns.heatmap(conf,annot=True,cmap=plt.cm.coolwarm)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

```python
%%writefile Model_Dep_app.py
import streamlit as st
import pickle
import pandas as pd
model=pickle.load(open('logistic_regression.pkl','rb'))
def user_input_parameters():
    Pclass =st.sidebar.selectbox("Select Passenger Class (Pclass)",[1,2,3])
    Sex= st.sidebar.selectbox("Select Your Gender,Female-1,Male-0",[0,1])
    Age= st.sidebar.slider('Select Your Age',0,100)
    SibSp = st.sidebar.slider("SibSp (siblings/spouses aboard)",0,10,0)
    Parch= st.sidebar.slider("Parch (Parents/Children aboard)",0,10,0)
    Fare= st.sidebar.number_input('Enter Fare',min_value=0.0,step=0.01)
    # Convert Embarked to numeric values
    embarked_options = {"C": 0, "Q": 1, "S": 2}  # Create a mapping for Embarked values
    embarked_display = st.sidebar.selectbox("Select Embarkation Port", ["C", "Q", "S"])
    Embarked = embarked_options[embarked_display]  # Convert to numeric value
    dict1={'Pclass':Pclass,'Sex': Sex,'Age':Age,'SibSp':SibSp,'Parch':Parch,'Fare':Fare,'Embarked':Embarked}
    features= pd.DataFrame(dict1,index=[0])
    return features
df= user_input_parameters()
pred_prob=model.predict_proba(df)
pred= model.predict(df)
st.subheader('Predicted')
st.write('Survived ' if pred_prob[0][1]>=0.5 else 'Not Survived ')
st.subheader('Predicted_Proba')
st.write(pred_prob)
```
