Skewness measures the asymmetry of a data distribution.Zero skewness: Symmetrical distribution(like a normal bell curve),Positive skewness(right-skewness):Tail is longer on the right & most data is on the left,Negative skewness(left-skewness):Tail is longer on the left side & most data is on the right.Formula of skewness:3(mean-median)/std.dev.

It gives the total degree of outliers present in the data.Kurtosis tells about how heavy or light the tails of a distribution are if High kurtosis means more extreme outliers(heavy tails) and low kurtosis means fewer outliers(light tails).

## Boxplots for each numerical columns

Boxplots is used to visualize the presence of outliers.The lower extreme and upper extreme the boundaries,any data lying outside the boundry is called outliers. It has first Quartile(Q1--25%th),median(Q2--50%th),third Quartile(Q3--75%th).The range in between Q1 and Q3 are called inter quantile range(IQR).Lower_extreme = Q1-1.5*IQR , Upper_extreme = Q3+1.5*IQR.

## Bar Charts for analysis of Catagorical columns

A bar chart used to represent categorical data with rectangular bars.It displays data using bars of different heights or lengths.Each bar represents a category, and its height shows the value or frequency

```python
df[cat_col]
```

```python
# Function to create bar charts for categorical columns
plt.figure(figsize=(8,5))    
sns.set(style='white')
plt.suptitle('Bar chart of categorical columns',color='red',fontsize=20,fontweight='bold')
def plot_categorical_distribution(df, col):
    df[col].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.tight_layout()
    plt.show();
# Generate bar charts for the identified categorical columns
categorical_columns = ['Date','Day', 'SKU', 'City', 'BU', 'Brand', 'Model']
for col in categorical_columns:
    plot_categorical_distribution(df, col)
```

From the above bar graph-- Date,SKU,BU and Model columns all the data has same number of count or evenly distributed. Where as in day and Brand columns----the count of data are different. 

## Standardization of Numerical Variables

Standardization or scaling: it is to scale the data into such a format that the mean become zero and the standard deviation become one.Standardization it should be in same scale(-3 to 3). Formula standardization--calculating the Z_score, Z_score = (data(x)-mean(u))/standard_deviation(sigma)

```python
df1=df[num_col]
df1
```

```python
# Creating an empty dataframe to store the standardize data
scaled_num_col=pd.DataFrame(columns=['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value'])
```

```python
scaled_num_col
```

```python
# Calculating the standardize data (Z_score)
scaled_num_col['Volume']=(df1['Volume']-df1['Volume'].mean())/df1['Volume'].std()
scaled_num_col['Avg Price']=(df1['Avg Price']-df1['Avg Price'].mean())/df1['Avg Price'].std()
scaled_num_col['Total Sales Value']=(df1['Total Sales Value']-df1['Total Sales Value'].mean())/df1['Total Sales Value'].std()
scaled_num_col['Discount Rate (%)']=(df1['Discount Rate (%)']-df1['Discount Rate (%)'].mean())/df1['Discount Rate (%)'].std()
scaled_num_col['Discount Amount']=(df1['Discount Amount']-df1['Discount Amount'].mean())/df1['Discount Amount'].std()
scaled_num_col['Net Sales Value']=(df1['Net Sales Value']-df1['Net Sales Value'].mean())/df1['Net Sales Value'].std()
```

```python
scaled_num_col
```

```python
scaled_column_mean=scaled_num_col.mean() ### Mean become Zero
scaled_column_mean
```

```python
scaled_column_std=scaled_num_col.std() ### Standard deviation become one
scaled_column_std
```

Standardization: mean become 0 and standard deviation become 1 and the range of standardize data lies in between -3 to 3 in all the columns.
As before all the numerical columns are in different scale but now it become same scale using standardization it easier to use and analyze.

## Conversion of Categorical Data into Dummy Variables

```python
## Scaling----Numerical----continous(features not in target column)
## Labeling---Categorical---discrete/text data
```

Most machine learning algorithms cannot handle categorical variables directly. One-hot encoding converts each category into a binary feature (0 or 1), making it easier for algorithms to interpret and process the data.The converted catagorical data are called dummy variables. If the data is nominal type, then One Hot Encoder is used to convert the data into binary format (0 or 1). If the data is ordinal, label encoder or Ordinal encoder is used to label the data by assigning a number to each catagory. Also we can use get_dummies in pandas package to get the dummy variables.

```python
df2=df[cat_col]
df2
```

```python
# Importing packages---- for converting
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
```

```python
# Using get_dummies function of pandas package
data=pd.get_dummies(df2,dtype='int')
data
```

```python
# Creating variable of functions to make is easy for using multiple times.
one_hot = make_pipeline(OneHotEncoder())
ord_enc = make_pipeline(OrdinalEncoder())
```

```python
# Using one hot encoder 
transformers=ColumnTransformer([('one_hot',one_hot,['BU'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.tail()
```

```python
transformers=ColumnTransformer([('one_hot',one_hot,['City'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
# Using ordinal encoder
transformers=ColumnTransformer([('ord_enc',ord_enc,['Brand'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
transformers=ColumnTransformer([('ord_enc',ord_enc,['Model'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```
